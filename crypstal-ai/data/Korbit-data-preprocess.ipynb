```python
balance_list = []
for idx, action in enumerate(action_list):
    validity = True
    if action == 0:
        if balance < df.close.values[idx]:
            validity = False
        if validity:
            balance = balance - df.close.values[idx]
            invest_amount = df.close.values[idx]
            balance -= invest_amount  # 보유 현금을 갱신
            num_coins += 1  # 보유 코인 수를 갱신
            buy += 1  # 매수 횟수 증가
            balance_list.append(balance)

    elif action == 1:
        if num_coins <= 0:
            validity = False
        if validity:
            invest_amount = df.close.values[idx]
            num_coins -= 1  # 보유 코인 수를 갱신
            balance += invest_amount  # 보유 현금을 갱신
            sell += 1  # 매도 횟수 증가
            balance_list.append(balance)
    if validity == False:
        hold += 1

        balance_list.append(balance)
```
